```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

```python
df=pd.read_csv('iris.csv')
```

```python
df.head()
```

```python
df.drop(['species'],axis=1,inplace=True)
df.head()
```

```python
from sklearn.cluster import KMeans
model = KMeans(n_clusters=3)
```

```python
model.fit(df)
```

```python
model.labels_
```

```python
from scipy.spatial.distance import cdist
distortions = []
K = range(1, 20)

for k in K:
    model = KMeans(n_clusters=k).fit(df)
    distortions.append(sum(np.min(cdist(df, model.cluster_centers_, 'euclidean'), axis=1)))
plt.plot(K, distortions, 'bx-')
plt.xlabel('k')
plt.ylabel('Distortion')
plt.title('The Elbow Method showing the optimal k')
plt.show()
```

```python
km=KMeans(n_clusters=3)
km.fit(df)
ym=km.predict(df)
center=km.cluster_centers_
km.labels_
```

```python
tdf =pd.DataFrame(data=center,columns=df.columns)
```

```python
tdf
```

```python
com_df=pd.concat([df,tdf])
```

```python
from sklearn.preprocessing import StandardScaler
df_norm=StandardScaler().fit_transform(com_df)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
trans_pc = pca.fit_transform(df_norm)
```

```python
trans_pc[-3:].shape
```

```python
trans_pc[-3:]
```

```python
pc_df = pd.DataFrame(data=trans_pc[:-3], columns=['pc1', 'pc2'])
pc_df['Cluster'] = km.labels_
```

```python
pc_df.head()
```

```python
first=pc_df[pc_df.Cluster==0]
second=pc_df[pc_df.Cluster==1]
third=pc_df[pc_df.Cluster==2]
```

```python
center_df=pd.DataFrame(data=trans_pc[-3:],columns =['pc1','pc2'])
center_df
```

```python
plt.scatter(x=first.pc1,y=first.pc2,color='blue',label='Cluster 0')
plt.scatter(x=second.pc1,y=second.pc2,color='purple',label='Cluster 1')
plt.scatter(x=third.pc1,y=third.pc2,color='green',label='Cluster 2')
plt.scatter(x=center_df.pc1,y=center_df.pc2,color='red',label='Centers')
plt.legend()
plt.show()
```
